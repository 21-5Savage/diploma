```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("xgboost_test_predictions.csv")
```

```python
ticker = "MSFT"
df_plot = df[df["ticker"] == ticker].sort_values("date")

fig = px.line(
    df_plot,
    x="date",
    y=["y_true", "y_pred"],
    title=f"{ticker} Prediction"
)

fig.show()
```
